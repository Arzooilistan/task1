```python
# ======================================================
# Multimodal Housing Price Prediction (Images + Tabular)
# Google Colab Compatible
# ======================================================

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# ------------------------------------------------------
# 1. Simulate Tabular + Image Dataset
# ------------------------------------------------------
num_samples = 500
num_features = 10
img_size = (3, 64, 64)

np.random.seed(42)
torch.manual_seed(42)

# Fake tabular data
tabular_data = np.random.rand(num_samples, num_features)
# Fake images
images = np.random.rand(num_samples, *img_size)
# Fake target prices
prices = np.random.randint(100000, 500000, size=num_samples)

# Train-Test Split
X_tab_train, X_tab_test, X_img_train, X_img_test, y_train, y_test = train_test_split(
    tabular_data, images, prices, test_size=0.2, random_state=42
)

# Standardize tabular features
scaler = StandardScaler()
X_tab_train = scaler.fit_transform(X_tab_train)
X_tab_test = scaler.transform(X_tab_test)

# ------------------------------------------------------
# 2. Dataset Class
# ------------------------------------------------------
class HousingDataset(Dataset):
    def __init__(self, tabular, images, targets):
        self.tabular = torch.tensor(tabular, dtype=torch.float32)
        self.images = torch.tensor(images, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.float32)

    def __len__(self):
        return len(self.targets)

    def __getitem__(self, idx):
        return self.tabular[idx], self.images[idx], self.targets[idx]

train_dataset = HousingDataset(X_tab_train, X_img_train, y_train)
test_dataset = HousingDataset(X_tab_test, X_img_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# ------------------------------------------------------
# 3. Define Model (CNN + Tabular MLP)
# ------------------------------------------------------
class MultimodalModel(nn.Module):
    def __init__(self, tabular_dim, hidden_dim=128):
        super(MultimodalModel, self).__init__()
        # CNN for image features
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))
        )
        # MLP for tabular data
        self.tabular_fc = nn.Sequential(
            nn.Linear(tabular_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2)
        )
        # Fusion
        self.fc = nn.Sequential(
            nn.Linear(32*4*4 + hidden_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 1)
        )

    def forward(self, tabular, images):
        img_features = self.cnn(images)
        img_features = img_features.view(img_features.size(0), -1)
        tab_features = self.tabular_fc(tabular)
        combined = torch.cat([img_features, tab_features], dim=1)
        output = self.fc(combined)
        return output.squeeze()

# ------------------------------------------------------
# 4. Training Setup
# ------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultimodalModel(tabular_dim=num_features).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ------------------------------------------------------
# 5. Training Loop
# ------------------------------------------------------
epochs = 5
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for tab, img, target in train_loader:
        tab, img, target = tab.to(device), img.to(device), target.to(device)
        optimizer.zero_grad()
        outputs = model(tab, img)
        loss = criterion(outputs, target)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * tab.size(0)
    print(f"Epoch [{epoch+1}/{epochs}] - Train Loss: {total_loss/len(train_loader.dataset):.2f}")

# ------------------------------------------------------
# 6. Evaluation
# ------------------------------------------------------
model.eval()
all_preds, all_targets = [], []
with torch.no_grad():
    for tab, img, target in test_loader:
        tab, img, target = tab.to(device), img.to(device), target.to(device)
        outputs = model(tab, img)
        all_preds.extend(outputs.cpu().numpy())
        all_targets.extend(target.cpu().numpy())

# Compute Metrics
mae = mean_absolute_error(all_targets, all_preds)
rmse = np.sqrt(mean_squared_error(all_targets, all_preds))   # ✅ Manual RMSE (Colab-safe)

print("\n✅ Evaluation Results:")
print(f"MAE : {mae:.2f}")
print(f"RMSE: {rmse:.2f}")

# ------------------------------------------------------
# 7. Visualization
# ------------------------------------------------------
plt.scatter(all_targets, all_preds, alpha=0.6)
plt.xlabel("True Prices")
plt.ylabel("Predicted Prices")
plt.title("True vs Predicted Housing Prices")
plt.show()
```
