```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visual style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# 1. Load dataset
print("Loading Iris dataset...")
iris = sns.load_dataset('iris')

# 2. Initial inspection
print("\n=== Dataset Shape ===")
print(f"Rows: {iris.shape[0]}, Columns: {iris.shape[1]}")

print("\n=== Column Names ===")
print(iris.columns.tolist())

print("\n=== First 5 Rows ===")
print(iris.head())

# 3. Summary statistics
print("\n=== Dataset Information ===")
print(iris.info())

print("\n=== Statistical Summary ===")
print(iris.describe())

# 4. Visualization
# Create figure with subplots
fig = plt.figure(figsize=(18, 15))

# 4a. Scatter plot - Relationships between features
print("\nGenerating visualizations...")
ax1 = fig.add_subplot(2, 2, 1)
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species',
    style='species',
    s=100,
    ax=ax1
)
ax1.set_title('Sepal vs Petal Length by Species')

# 4b. Histograms - Feature distributions
ax2 = fig.add_subplot(2, 2, 2)
sns.histplot(
    data=iris,
    x='sepal_width',
    hue='species',
    kde=True,
    element='step',
    ax=ax2
)
ax2.set_title('Sepal Width Distribution')

# 4c. Box plots - Outlier detection
ax3 = fig.add_subplot(2, 2, 3)
sns.boxplot(
    data=iris,
    x='species',
    y='petal_width',
    hue='species',
    dodge=False,
    ax=ax3
)
ax3.set_title('Petal Width Distribution by Species')


# 4d. Pair plot - Comprehensive relationships
ax4 = fig.add_subplot(2, 2, 4)
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('Feature Pair Relationships', y=1.02)

# Adjust layout and display
plt.tight_layout()
plt.show()

print("\n=== Analysis Complete ===")
```
