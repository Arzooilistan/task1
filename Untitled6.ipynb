```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# Task objective
"""
Predict the next day's closing price for Apple (AAPL) using:
  - Historical features: Open, High, Low, Volume
  - Model: Random Forest Regressor
"""

# Dataset: Apple (AAPL) stock data from 2020-01-01 to 2024-12-31
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")

# Preprocessing
def preprocess_data(df):
    df = df.copy()
    # Create target: next day's closing price
    df["Target"] = df["Close"].shift(-1)
    # Remove last row (NaN target)
    df = df[:-1]
    return df

df = preprocess_data(data)

# Features and target
features = ["Open", "High", "Low", "Volume"]
X = df[features]
y = df["Target"]

# Split data chronologically (80% train, 20% test)
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# Initialize and train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluate model
print(f"Train R²: {r2_score(y_train, train_preds):.4f}")
print(f"Test R²: {r2_score(y_test, test_preds):.4f}")
print(f"Test MAE: ${mean_absolute_error(y_test, test_preds):.2f}")

# Plot predictions vs actual
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label="Actual Price", color="blue")
plt.plot(y_test.index, test_preds, label="Predicted Price", color="red", linestyle="--")
plt.title("Apple Stock: Actual vs Predicted Closing Price")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```
